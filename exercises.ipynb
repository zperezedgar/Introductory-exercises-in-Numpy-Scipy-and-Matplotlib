1-Create the matrices (matrix or ndarray)


*   The matrix S whose entries are aij=i+j
*  The matrix P whose entries are aij=i*j
*  The matrices A,B,C







```
import numpy as np

def S(dim):
  s=np.zeros((dim,dim))
  for i in range(dim):
    for j in range(dim):
      s[i,j]=i+j+2
  return s

def P(dim):
  p=np.zeros((dim,dim))
  for i in range(dim):
    for j in range(dim):
      p[i,j]=(i+1)*(j+1)
  return p

A=np.zeros((4,4))
a=0
for i in range(4):
  for j in range(4):
    a+=1
    A[i,j]=a

B=np.zeros((4,4))
b=0
for i in range(4):
  for j in range(4):
    b+=1
    B[i,j]=b
    if(i==j):
      B[i,j]=0

C=np.zeros((4,4))
c=0
for i in range(4):
  for j in range(4):
    c+=1
    C[i,j]=c
    if(i==j):
      C[i,j]=(i+1)*(j+1)

print(S(3))
print(P(3))
print(A)
print(B)
print(C)
```

5-Minimize the following functions (from scipy import optimize)

```
from scipy import optimize
import numpy as np

def f1(x):
  return 3*(x**4)+(x+1)**2

def f2(x):
  return x+np.cos(x**2)

def f3(x):
  return (4-2.1*(x[0]**2)+(x[0]**4)/3)*(x[0]**2) +x[0]*x[1]+((4*x[1]**2)-2)*(x[1]**2)
```

```
optimize.minimize(f1,0)
```

```
optimize.minimize(f1,2)
```

```
optimize.minimize(f2,0)
```

```
optimize.minimize(f2,-15)
```

```
x0=np.array([0,0.5])
optimize.minimize(f3,x0)
```

6-Use the function optimize.curve:fit() to fix the functions. Note: First generate a set of points (x,y)=(x+f(x)+n) where n is a gaussian noise

```
import matplotlib.pyplot as plt
import numpy as np
import math

def fx1(a,b,c,x):
  return a*(x**2)+b*x+c

def fx2(a,b,c,x):
  return a*(x**2)+b*np.sin(x)

def fx3(a,b,c,x):
  return a*(np.exp(-b*x))+c

def fx4(a,b,c,x):
  return a*(x**3)+b*x

def fx5(a,b,c,x):
  return a*(x**2)+b*x+c*(np.exp(x))

a,b,c=1.1,2.3,0.4
mu, sigma = 0.1, 0.1 # mean and standard deviation
x=np.linspace(0,10,20)

np.random.seed(1672)
y1=fx1(a,b,c,x)+np.random.normal(mu,sigma)*1
y2=fx2(a,b,c,x)+np.random.normal(mu,sigma)*1
y3=fx3(a,b,c,x)+np.random.normal(mu,sigma)*1
y4=fx4(a,b,c,x)+np.random.normal(mu,sigma)*1
y5=fx5(a,b,c,x)+np.random.normal(mu,sigma)*1

ctesfit1,cov1=optimize.curve_fit(fx1,x,y1,[a,b,c])
plt.scatter(x,y1)
plt.plot(x,fx1(a,b,c,x),x,fx1(*ctesfit1,x))
plt.xlabel('x')
plt.ylabel('y')
```

```
ctesfit2,cov2=optimize.curve_fit(fx2,x,y2)
plt.scatter(x,y2)
plt.plot(x,fx2(a,b,c,x),x,fx2(*ctesfit2,x))
plt.xlabel('x')
plt.ylabel('y')
```

```
ctesfit3,cov3=optimize.curve_fit(fx3,x,y3)
plt.scatter(x,y3)
plt.plot(x,fx3(a,b,c,x),x,fx3(*ctesfit3,x))
plt.xlabel('x')
plt.ylabel('y')
```

```
ctesfit4,cov4=optimize.curve_fit(fx4,x,y4)
plt.scatter(x,y4)
plt.plot(x,fx4(a,b,c,x),x,fx4(*ctesfit4,x))
plt.xlabel('x')
plt.ylabel('y')
```

```
ctesfit5,cov5=optimize.curve_fit(fx5,x,y5)
plt.scatter(x,y5)
plt.plot(x,fx5(a,b,c,x),x,fx5(*ctesfit5,x))
plt.xlabel('x')
plt.ylabel('y')
```
