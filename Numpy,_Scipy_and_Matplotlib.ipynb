**NUMPY PACKAGE**

Manual construction of array 1D:

```
import numpy as np

a=np.array([1,21,3])
print(a.ndim)
print(a.shape)
print(a.size)
print(a.max())
print(a.argmax()) #Returns the indices of the maximum values along an axis.
```

Manual Construction of array 2D:

```
import numpy as np

a=np.array([[1,21,3],[2,44,6]])
print(a.ndim)
print(a.shape)
print(a.size)
print(a.max())
print(a.argmax()) #Returns the indices of the maximum values along an axis.
```

Manual construction of an array 3D:

```
import numpy as np

a=np.array([[[1,21,3],[2,44,6]],[[1,2,1],[20,4,2]]])
print(a.ndim)
print(a.shape)
print(a.size)
print(a.max())
print(a.argmax()) #Returns the indices of the maximum values along an axis.
```

Functions for creating arrays:

```
print(np.arange(10)) #0..n-1
print(np.arange(1,9,2)) #start,end(exclusive),step
print(np.linspace(0,1,6)) #start,end,num-points
print(np.linspace(0,1,5,endpoint=False))
print(np.ones((3,3))) #(3,3) is a tuple
print(np.zeros((3,3)) )
print(np.eye(3))
print(np.diag(np.array([1,2,3,4])))
print(np.random.rand(4,3)) #uniform in [0,1]
print(np.random.randn(2,5)) #Gaussian
print(np.random.seed(1234)) #Setting the random seed
```

Basic Data types:

```
import numpy as np
a=np.array([[1.,21.,3.],[2,44,6]],np.int8)
b=np.array([[1.,21.,3.],[2,44,6]],np.uint64)

print(a.ndim)
print(a.shape)
print(a.dtype)
print(b.dtype)
```

Indexing, slicing and assignment 1D.
The items of an array can be accessed and assigned to the same way as other Python sequences (e.g.lists):

One dimensional arrays:

```
a=np.arange(10)
print(a[0])
print(a[-1]) #a[len(a)-1]
print(a[::2])
print(a[::-1])
a[0]=10
print(a)
a[0:2]=[12,16] #copia la sequencia de longitud dada en el Ã­ndice dado
print(a)
```

Deleting and inserting 1D.

```
#a[0]=[] Error!!!
a=np.arange(10)
print(a)
a=np.delete(a,0) #delete the first element
print(a)
a=np.delete(a,-1) #delete the last element
print(a)
a=np.insert(a,0,5) #array, index, value
print(a)
#insert the value before the index
```

Multi dimensional arrays:

```
import numpy as np

a=(10*np.random.rand(5,5)).astype(int)
print(a)
print(a[0,0])
print(a[:,1])
print(a[::2,::2])
print(a[0:2,1:4])
print(a[1:4,3:5]) #a[1:4,:][:,3:5]
print(a[np.ix_([1,3,4],[0,2])]) #This function takes N 1-D sequences and returns
#N outputs with N dimensions each, such that the shape is 1 in all but one dimension 
#and the dimension with the non-unit shape value cycles through all N dimensions.
```

```
import numpy as np

a=(10*np.random.rand(5,5)).astype(int)
print(a)
a[0,0]*=10
print(a)
a[1:4,3:5]=100*np.ones((3,2)) #a[1:4,3:5]=100
print(a)
#a[np.ix_([1,3],[0,4])]=np.random.rand((3,2))
#print(a)
print((a>4).choose(a,4))
print(a.clip(min=None,max=4)) #rescalar
print(a.clip(min=3,max=7)) #rescalar
a[:]=3
print(a)
```

Deleting and inserting in multi dimensional arrays:

```
a=(10*np.random.rand(5,5)).astype(int)
print(a)
print(np.delete(a,0,0)) #arr,obj(int),axis
print(np.delete(a,np.s_[::2],0)) #arr,obj(slice),axis
print(np.delete(a,np.arange(0,4,2),0)) #obj (array)
print(np.delete(a,[1,3])) #np.delete(a,[1,3],None)
print(np.delete(a,[1,3],0))
print(np.insert(a,1,100,1)) #arr,obj(int,slice,seq), #value,axis
print(np.insert(a,(1,4),100,1))
print(np.insert(a,1,[1,2,3,4,5],1))
ia=np.zeros((5,2))
print(ia)
print(np.insert(a,(1,4),ia,1))
```

```
a=(10*np.random.rand(5,5)).astype(int)
print(a)
print(a.take([0,2,2], axis=0).take([2,4],axis=1))
print(a.diagonal(offset=0)) #a.diagonal(0)
print(a.sum(axis=0)) #a.sum(0)
print(a.sum(axis=1)) #a.sum(1)
print(a.sum())
print(a.trace(offset=0)) #a.trace(0)
print(a.max()) #a.max(0), a.max(1)
print(a.T) #a.transpose()
```

Convert ND array into a vector (column or row):

```
a=(10*np.random.rand(5,5)).astype(int)
print(a)
#convert into an 1D array
c=a.flatten()
print(c)
#column vector
c=a.flatten()[:,np.newaxis]
print(c)
#column vector
c=np.array([2,3,4,5]).reshape(-1,1)
print(c)
```

Concatenation

```
a=np.array([[0,1,2],[3,4,5]])
print(a)
c=np.concatenate((a,a),0) #np.vstack((a,a))
print(c)
c=np.concatenate((a,a),1) #np.hstack((a,a))
print(c)
```

NumPy provides many functions for perfoming standard vector and matrix multiplication routines

```
a=np.array([1,2,3],float)
print(a)
b=np.array([0,1,1],float)
print(b)
print(np.dot(a,b))
print(np.inner(a,b))
print(np.outer(a,b))
```

Linear Algebra

```
a=np.array([[4,2,0],[9,3,7],[1,2,1]])
print(a)
b=np.linalg.inv(a)
print(b)
print(np.dot(a,b))
U,D,V=np.linalg.svd(a)
np.dot(np.dot(U,np.diag(D)),V) #==a
```

Universal functions:
A universal function performs elementwise operations on data in ndarrays.

```
a=np.array([1,2,3])
np.sqrt(a)
```

**Matplotlib**

```
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-1,1,100); y=x**2
plt.plot(x,y)
plt.show(False)
```

Adding graphics in the same figure

```
import numpy as np 

x=np.linspace(-np.pi,np.pi,100);
ys,yc,y2=np.sin(x),np.cos(x),x**2

plt.plot(x,ys)
plt.plot(x,yc)
plt.plot(x,y2)
```

or simply

```
plt.plot(x,ys,x,yc,x,y2)
```

Adding style

```
#Create a new figure of size 8x6 inches, using 100 dpi
plt.figure(1,figsize=(8,6),dpi=100) #num,figsize,dpi
x=np.linspace(-np.pi,np.pi,100);
s,c=np.sin(x),np.cos(x)
plt.plot(x,s,color="blue",linewidth=1.0)
```

```
plt.plot(x,s,color="green",linewidth=1.0)
```

```
#Set x,y limits
plt.xlim(-4,4); plt.ylim(-1,1)
plt.plot(x,s,color="green",linewidth=1.0)
```

```
#Set x,y thicks
plt.xticks(np.linspace(-4,4,9)),plt.yticks(np.linspace(-4,4,9))
plt.plot(x,s,color="green",linewidth=1.0)
```

```
#Save figure using 80 dots per inch
plt.plot(x,s,color="green",linewidth=1.0)
plt.savefig("excercice1.png",dpi=80)
```

Example subplot

```
plt.figure(1,figsize=(8,6),dpi=100) #num,figsize,dpi
x=np.linspace(-np.pi,np.pi,100);
s,c,t,y2=np.sin(x),np.cos(x),np.tan(x),x**2
plt.subplot(2,2,1);plt.plot(x,s,'r-')
plt.subplot(2,2,2);plt.plot(x,c,'b-')
plt.subplot(2,2,3);plt.plot(x,t,'g-')
plt.subplot(2,2,4);plt.plot(x,y2,'k-')
```
